```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from pysal.viz import mapclassify
import seaborn as sns
import numpy as np
import contextily as cx
```

### Task 1: COVID-19 choropleths

1. Create 3-4 different choropleths using classification methods your own choice based on the number of covid cases per 100,000 people (column **Cumulative_incidens_per_100000**), and complement them with a figure that displays the distribution of values using a KDE.

2. Create a choropleth map with qualitative values based on the following method:
- Split the municipalities in two classes: above and below the median COVID-19 rate
- Assign a qualitative label (`above` or `below`) to each municipality
- Create a unique value map for the labels you have just created

```python
cc = pd.read_csv('data2023/covid_cases.csv')
mun = gpd.read_file('data2023/municipalities_dk.gpkg')
```

```python
gdf = mun.merge(cc, on='municipal_id')
col = 'Cumulative_incidens_per_100000'
```

### Equal interval

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))

k = 5

# plot 1
gdf.plot(
    column=col,
    scheme='EqualInterval',
    k=k,
    legend=True,
    ax=ax1
)

# plot 2
eq = mapclassify.EqualInterval(gdf[col], k=k)
sns.kdeplot(gdf[col], fill=True, ax=ax2)
sns.rugplot(gdf[col], alpha=0.5, ax=ax2)
for cut in eq.bins:
    ax2.axvline(cut, color='red', linewidth=0.75)
plt.tight_layout()
```

### Quantile plot

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))

# plot 1
gdf.plot(
    column=col,
    scheme='quantiles',
    k=k,
    legend=True,
    ax=ax1
)

# plot 2
qu = mapclassify.Quantiles(gdf[col], k=k)
sns.kdeplot(gdf[col], fill=True, ax=ax2)
sns.rugplot(gdf[col], alpha=0.5, ax=ax2)
for cut in qu.bins:
    ax2.axvline(cut, color='red', linewidth=0.75)
plt.tight_layout()
```

### fisher_jenks

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))

# plot 1
gdf.plot(
    column=col,
    scheme='fisher_jenks',
    k=k,
    legend=True,
    ax=ax1
)

# plot 2
fj = mapclassify.FisherJenks(gdf[col], k=k)
sns.kdeplot(gdf[col], fill=True, ax=ax2)
sns.rugplot(gdf[col], alpha=0.5, ax=ax2)
for cut in fj.bins:
    ax2.axvline(cut, color='red', linewidth=0.75)
plt.tight_layout()
```

### qualitative


```python
gdf['qual'] = (gdf[col] >= gdf[col].median()).map({True: 'Above', False: 'Below'})
gdf.plot(
    column='qual',
    categorical=True,
    legend=True,
    cmap='coolwarm'
)
```

### Task 2: Compare Classification Schemes

Evaluate the classification schemes (not including the last, manual one) using the approach presented [here](https://geographicdata.science/book/notebooks/05_choropleth.html#comparing-classification-schemes).

**Which one would you use? Why?**

```python
# Bunch classifier objects
class5 = eq, qu, fj
# Collect ADCM for each classifier
fits = np.array([c.adcm for c in class5])
# Convert ADCM scores to a DataFrame
adcms = pd.DataFrame(fits)
# Add classifier names
adcms["classifier"] = [c.name for c in class5]
# Add column names to the ADCM
adcms.columns = ["ADCM", "Classifier"]
ax = sns.barplot(
    y="Classifier", x="ADCM", data=adcms, palette="Pastel1"
)
```
